# PyCity Schools Analysis

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Gets a unique count of the schools in the district
dist_total_school = school_data_complete["school_name"].nunique()

#Gets a count of all the students in the district
dist_total_students = school_data_complete["Student ID"].count()

#Gets a total of the district's budget by summing only uniques 
dist_total_budget = school_data_complete["budget"].unique().sum()

#Gets the district's average math score
dist_avg_math_score = school_data_complete["math_score"].mean()

#Gets the district's average reading score
dist_avg_reading_score = school_data_complete["reading_score"].mean()

#Gets the overall district average
dist_overall_pass_rate = (dist_avg_math_score + dist_avg_reading_score)/2

#Calculates the math passing percentage 
dist_math_student_pass = school_data_complete.loc[school_data_complete["math_score"] > 69,:]
dist_math_student_pass_count = dist_math_student_pass["Student ID"].count()
dist_math_student_pass_percent = (dist_math_student_pass_count/dist_total_students)*100

#Calculates the reading passing percentage
dist_read_student_pass = school_data_complete.loc[school_data_complete["reading_score"] > 69,:]
dist_read_student_pass_count = dist_read_student_pass["Student ID"].count()
dist_read_student_pass_percent = (dist_read_student_pass_count/dist_total_students)*100

#create the summary dataframe
district_summary_df = pd.DataFrame({"Total Schools":dist_total_school,
                                    "Total Students":dist_total_students,
                                    "Total Budget":dist_total_budget,
                                    "Average Math Score":dist_avg_math_score,
                                    "Average Reading Score":dist_avg_reading_score,
                                    "% Passing Math":dist_math_student_pass_percent,
                                    "% Passing Reading":dist_read_student_pass_percent,
                                    "% Overall Passing":[dist_overall_pass_rate]
                                   })

#Add formatting to clean up the dataframe
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Gets the type for each school
sch_sumry_school_type = school_data_complete.groupby("school_name")["type"].first()

#Gets the total students per school
sch_sumry_total_students = school_data_complete.groupby("school_name")["size"].first()

#gets the budget for each school
sch_sumry_total_school_budget = school_data_complete.groupby("school_name")["budget"].first()

#Calculates the budget per student at each school
sch_sumry_per_student_budget = sch_sumry_total_school_budget/sch_sumry_total_students

#Gets the average math score for each school
sch_sumry_avg_math_score = school_data_complete.groupby("school_name")["math_score"].mean()

#Gets the average reading score for each school
sch_sumry_avg_read_score = school_data_complete.groupby("school_name")["reading_score"].mean()

#Gets the number of students who passed math and calculates the percentage for each school
sch_sumry_passing_math_filtered_df = school_data_complete[school_data_complete.math_score > 69]
sch_sumry_passing_math_count = sch_sumry_passing_math_filtered_df.groupby("school_name")["Student ID"].count()
sch_sumry_percent_passing_math = (sch_sumry_passing_math_count/sch_sumry_total_students)*100

#Gets the number of students who passed reading and calculates the percentage for each school 
sch_sumry_passing_read_filtered_df = school_data_complete[school_data_complete.reading_score > 69]
sch_sumry_passing_read_count = sch_sumry_passing_read_filtered_df.groupby("school_name")["Student ID"].count()
sch_sumry_percent_passing_read = (sch_sumry_passing_read_count/sch_sumry_total_students)*100

#calculates the overall passing rate for each school
sch_sumry_overall_pass_rate = (sch_sumry_percent_passing_math + sch_sumry_percent_passing_read)/2

#Create dataframe for School Summary
school_summary_df = pd.DataFrame({"School Type":sch_sumry_school_type,
                                  "Total Students":sch_sumry_total_students,
                                  "Total School Budget":sch_sumry_total_school_budget,
                                  "Per Student Budget":sch_sumry_per_student_budget,
                                  "Average Math Score":sch_sumry_avg_math_score,
                                  "Average Reading Score":sch_sumry_avg_read_score,
                                  "% Passing Math":sch_sumry_percent_passing_math,
                                  "% Passing Reading":sch_sumry_percent_passing_read,
                                  "Overall Passing Rate":sch_sumry_overall_pass_rate})
school_summary_df.head()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#Sorts the summary by the school with the highest passing rate
school_summary_top_perf_df = school_summary_df.sort_values(by=["Overall Passing Rate"], ascending=False)
school_summary_top_perf_df.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
#Sorts the summary by the lowest performing school. 
school_summary_bottom_perf_df = school_summary_df.sort_values(by=["Overall Passing Rate"], ascending=True)
school_summary_bottom_perf_df.head()
```

## Math Scores by Grade

* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#extracts only the 9th grade data and gets the average math grade for the 9th graders in each school. 
math_grade_9_df = school_data_complete.loc[(school_data_complete["grade"] == "9th"),:]
math_grade_9_school = math_grade_9_df.groupby("school_name")["math_score"].mean()

#extracts only the 10th grade data and gets the average math grade for the 10th graders in each school. 
math_grade_10_df = school_data_complete.loc[(school_data_complete["grade"] == "10th"),:]
math_grade_10_school = math_grade_10_df.groupby("school_name")["math_score"].mean()

#extracts only the 11th grade data and gets the average math grade for the 11th graders in each school. 
math_grade_11_df = school_data_complete.loc[(school_data_complete["grade"] == "11th"),:]
math_grade_11_school = math_grade_11_df.groupby("school_name")["math_score"].mean()

#extracts only the 12th grade data and gets the average math grade for the 12th graders in each school. 
math_grade_12_df = school_data_complete.loc[(school_data_complete["grade"] == "12th"),:]
math_grade_12_school = math_grade_12_df.groupby("school_name")["math_score"].mean()

#create the dataframe
math_scores_df = pd.DataFrame({"9th":math_grade_9_school,
                               "10th":math_grade_10_school,
                               "11th":math_grade_11_school,
                               "12th":math_grade_12_school
                              })

#cleans up the results by rounding to the second decimal. 
math_scores_df.round(2).head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Bins and labels created for spending range.
spending_per_student_bins = [0, 585, 615, 645, 675]
spending_group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

#Added a column to the school_summary_df to categorize each school by the spending range
school_summary_df["Spending Range"] = pd.cut(school_summary_df["Per Student Budget"], spending_per_student_bins, labels=spending_group_names)

#Gets the average math score for each school.  And grouped by the spending range
spending_avg_math = school_summary_df.groupby("Spending Range")["Average Math Score"].mean()

#Gets the average reading score for each school.  And grouped by the spending range
spending_avg_read = school_summary_df.groupby("Spending Range")["Average Reading Score"].mean()

#Gets the average math passing rate for each school.  And grouped by the spending range
spending_pass_math = school_summary_df.groupby("Spending Range")["% Passing Math"].mean()

#Gets the average reading passing rate for each school.  And grouped by the spending range
spending_pass_read = school_summary_df.groupby("Spending Range")["% Passing Reading"].mean()

#calculates the overall passing rate for each range based on the values above.
spending_pass_overall = (spending_pass_math + spending_pass_read)/2

#create dataframe
school_summary_spending_df = pd.DataFrame({"Average Math Score":spending_avg_math, 
                                           "Average Reading Score":spending_avg_read, 
                                           "% Passing Math":spending_pass_math,
                                           "% Passing Reading":spending_pass_read, 
                                           "Overall Passing Rate":spending_pass_overall
                                          })


school_summary_spending_df
```

## Scores by School Size

```python
# bins for size and names for the size range created
size_bins = [0, 1000, 2000, 5000]
size_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
#add a new column to school_summary_df for size range
school_summary_df["Size Range"] = pd.cut(school_summary_df["Total Students"], size_bins, labels=size_names)

#gets the average math score for each school and groups them by the size range
size_avg_math = school_summary_df.groupby("Size Range")["Average Math Score"].mean()

#Gets the average reading score for each school and groups them by the size range
size_avg_read = school_summary_df.groupby("Size Range")["Average Reading Score"].mean()

#Gets the math passing rate and groups them by size range
size_pass_math = school_summary_df.groupby("Size Range")["% Passing Math"].mean()

#Gets the reading passing rate for each school and groups them by size range
size_pass_read = school_summary_df.groupby("Size Range")["% Passing Reading"].mean()

#calculates the overall passing rate for each school. 
size_pass_overall = (size_pass_math + size_pass_read)/2

#create dataframe
school_summary_size_df = pd.DataFrame({"Average Math Score":size_avg_math, 
                                       "Average Reading Score":size_avg_read, 
                                       "% Passing Math":size_pass_math,
                                       "% Passing Reading":size_pass_read, 
                                       "Overall Passing Rate":size_pass_overall
                                       })


school_summary_size_df

```
